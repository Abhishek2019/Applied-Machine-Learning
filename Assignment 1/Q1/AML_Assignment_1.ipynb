#### Linear Regression

```python
from sklearn.model_selection import KFold
from sklearn import linear_model
import math

target_col = 'Healthy life expectancy at birth'
kf = KFold(n_splits=4, random_state=None, shuffle=False)

for i, (train_index, val_index) in enumerate(kf.split(train)):

  print("iteration ------------- ",i)
  train_X = train.iloc[train_index][[i for i in train.columns if i != target_col]]
  train_y = train.iloc[train_index][target_col]

  val_X = train.iloc[val_index][[i for i in train.columns if i != target_col]]
  val_y = train.iloc[val_index][target_col]

  print(train_X.shape, train_y.shape,val_X.shape,val_y.shape)

  reg = linear_model.LinearRegression()
  reg.fit(train_X, train_y)

  pred = reg.predict(val_X)

  print("Regression Score : ", reg.score(val_X,val_y))

  rmse = math.sqrt( ( (np.array(pred)-np.array(val_y))**2).sum()/len(pred) )
  print("RMSE : ", rmse)





```
