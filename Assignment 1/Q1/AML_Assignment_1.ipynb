#### B] Display the statistical values for each of the attributes, along with visualizations (e.g., histogram) of the distributions for each attribute. Explain noticeable traits for key attributes.
```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

for col in data.columns:
  print("----------------------------------------")
  print(col)
  if col in categorical_attr:
    bins = list(data[col].value_counts().index)
    counts = list(data[col].value_counts())
    if len(bins) > 20:
      plt.figure(figsize=(40,5))
      plt.bar(bins,counts,ec='black')
      plt.xticks(rotation=90)
      plt.show()
    else:
      plt.figure(figsize=(10,5))
      plt.bar(bins,counts,ec='black')
      plt.xticks(rotation=90)
      plt.show()

  else:
      counts,bins = np.histogram(data[col])
      plt.figure(figsize=(10,5))
      plt.stairs(counts,bins,ec='black')
      plt.xticks(rotation=90)
      plt.show()
```

Since categorical labels are large in number, we need to implement one hot encoding on it

```python
processed_data = data.copy()
```

```python
processed_data.shape
```

```python
for col in categorical_attr:
  temp_dummy = pd.get_dummies(data[col])
  processed_data = processed_data.join(temp_dummy)
  processed_data = processed_data.drop([col],axis=1)
```
